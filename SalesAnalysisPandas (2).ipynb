 <h3>exploratary data analysis

<h4> what is the overall sales trend?

```python
df['order_date'].min()
```

```python
df['order_date'].max()
```

```python
#getting month year from the dataset
df['month_year'] = df['order_date'].apply(lambda x: x.strftime('%Y-%m'))
```

```python
#grouping month year
df_trend = df.groupby('month_year').sum(numeric_only=True)['sales'].reset_index()
```

```python

# setting fig size
plt.figure(figsize=(15,6))
plt.plot(df_trend['month_year'], df_trend['sales'])
plt.xticks(rotation='vertical', size=8)
plt.show()
```

<h4> which are the top products by sales?  

```python

#grouping product name column
prod_sales = pd.DataFrame(df.groupby('product_name')['sales'].sum(numeric_only=True))

#df_sum = df.groupby(['product_name', 'sales']).sum(float_only= True)
```

```python
#sorting prod_sales column
prod_sales = prod_sales.sort_values('sales',ascending= False)
```

```python
#top 10 by products by sales
prod_sales[:10]
```

<h4> which are the most selling product?

```python
#group product name 
most_selled = pd.DataFrame(df.groupby('product_name')['quantity'].sum())
```

```python
# sorting most_selled
most_selled = most_selled.sort_values('quantity',ascending=False)
```

```python
most_selled[:10]
```

<h4>which is the most preferred ship mode?

```python
#setting figure size 
plt.figure(figsize=(10,8.5))
sns.countplot(x='ship_mode', data=df)
plt.show()
```

<h4>which are the most profitable category and sub-category?

```python
cat_subcat_profit =pd.DataFrame(df.groupby(['category','sub_category']).sum(numeric_only=True)['profit'])
```

```python
cat_subcat_profit.sort_values(['category','profit'],ascending=False)  
```
